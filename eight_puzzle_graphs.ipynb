```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("puzzle_8_results - Copy.csv")
df.head()
```

# Plot for each algorithm and board 

```python
# fig, ax = plt.subplots(figsize=(8,8))
for curr_algorithm in ['BFS', 'A_star']:
    for curr_board_index in range(3):
        curr_df = df[(df.Algorithm == curr_algorithm) & (df.board_index == curr_board_index)]
        curr_df.plot(x='K', y='average_time')
        plt.ylabel("time (seconds)")
        plt.title(f"{curr_algorithm} Algorithm, Board index - {curr_board_index}")
        plt.plot()
```

# Compare between BFS and A* for each board in times


```python
width = 0.4

for curr_board_index in range(3):
    fig = plt.figure(figsize=(15,7))
    ax = fig.add_axes([0,0,1,1])
    
    for idx, curr_algorithm in enumerate(['BFS', 'A_star']):
        curr_df = df[(df.Algorithm == curr_algorithm) & (df.board_index == curr_board_index)]
        
        ax.bar(curr_df.K + width * idx, curr_df.average_time, width=width, label=curr_algorithm)
        plt.ylabel("time (seconds)")
    plt.legend()
    plt.title(f"Board index - {curr_board_index}")
    plt.plot()
        
```

# Time vs steps

```python

for curr_algorithm in ['BFS', 'A_star']:
    for curr_board_index in range(3):
        fig, ax1 = plt.subplots()
        curr_df = df[(df.Algorithm == curr_algorithm) & (df.board_index == curr_board_index)]
        
        ax1.set_xlabel("K")
        color = 'red'
        ax1.set_ylabel('time', color=color)
        ax1.plot(curr_df.K, curr_df.average_time, color = color)
        ax1.tick_params(axis='y', labelcolor=color)
        ax1.set_label('time')
        
        ax2 = ax1.twinx()
        color = 'blue'
        ax2.set_ylabel('steps', color=color)
        ax2.plot(curr_df.K, curr_df.total_steps, color=color)
        ax2.tick_params(axis='y', labelcolor=color)
        
        plt.title(f"{curr_algorithm} Algorithm, Board index - {curr_board_index}")
        plt.plot()
```
